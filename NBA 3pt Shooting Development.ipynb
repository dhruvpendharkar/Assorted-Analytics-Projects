```python
from sportsreference.nba.roster import Player
from sportsreference.nba.roster import Roster
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
# function to identify players on team

def team(abbr, year):
    roster = Roster(abbr, year)
    list1 = []
    for player in roster.players:
        list1.append(player.player_id)
        
    return list1

# function to get shooting data for all players careers

def get_data(id_list):
    player_data = {}
    for i in id_list:
        player = Player(i)
        df = player.dataframe
        three_point_score = []
        three_point_score = []
        for p in range(len(df["three_point_percentage"]) - 1):
            per = df["three_point_percentage"][p]
            att = df["three_point_attempts"][p]
            minutes = df["minutes_played"][p]
            if(minutes != 0 and att != 0 and per != 0):
                score = ((per * 10) ** 2) * ((att/minutes) * 36)
            else:
                score = 0
            three_point_score.append(score)
        
        player_data[player.name] = three_point_score
    return player_data


def gen_plots(dictionary):
    prof_list = list(dictionary.values())
    player_list = list(dictionary.keys())
    for z in range(len(prof_list)):
        l = prof_list[z]
        time = range(len(l))
        plt.scatter(time, l)
        plt.title(player_list[z])
        plt.xlabel("Season")
        plt.ylabel("Three point shooting proficiency")
        plt.show()
        
        
def gen_regr(dictionary):
    prof_list = list(dictionary.values())
    player_list = list(dictionary.keys())
    coef_list = []
    for z in range(len(prof_list)):
        l = prof_list[z]
        time = range(len(l))
        r = LinearRegression()
        r.fit([l], [time])
        coef_list.append(r.coef_)
        
    return coef_list

def get_data(player_id):
    player_data = {}
    player = Player(player_id)
    df = player.dataframe
    three_point_score = []
    three_point_score = []
    for p in range(len(df["three_point_percentage"]) - 1):
        per = df["three_point_percentage"][p]
        att = df["three_point_attempts"][p]
        minutes = df["minutes_played"][p]
        if(minutes != 0 and att != 0 and per != 0):
            score = ((per * 10) ** 2) * ((att/minutes) * 36)
        else:
            score = 0
        three_point_score.append(score)
        
        player_data[player.name] = three_point_score
        
    return player_data
```
