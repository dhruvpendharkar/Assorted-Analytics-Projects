```python
## Percentage of times that a player scores more than 2 standard deviations from their mean score
import numpy as np
def scoringVolatility(x):
    avg = np.mean(x)
    std = np.std(x)
    count = 0
    for i in x:
        if(i>avg + (2*std)):
            count += 1
        elif(i<avg-(2*std)):
            count += 1
            
    return (count/len(x)) * 100

```
