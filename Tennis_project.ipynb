```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

tennis_data = pd.read_csv(r"C:\Users\monik\Downloads\tennis_ace_starting\tennis_ace_starting\tennis_stats.csv")

return_points_w = tennis_data['ReturnPointsWon']
wins = tennis_data['Wins']
break_points = tennis_data['BreakPointsFaced']

features = tennis_data[['BreakPointsFaced']]
outcome = tennis_data[['Wins']]
features_train, features_test, outcome_train, outcome_test = train_test_split(features, outcome, train_size=0.8)
model1 = LinearRegression()
model1.fit(features_train, outcome_train)
model1.score(features_test, outcome_test)
predict = model1.predict(features_test)

plt.scatter(predict, outcome_test, alpha=0.4)
plt.title("Predicted Wins vs. Test Set")
plt.xlabel("Predicted Wins")
plt.ylabel("Test Set")
plt.show()

plt.scatter(wins, break_points, alpha=0.4)
plt.title("Break Points Faced vs. Wins")
plt.xlabel("Wins")
plt.ylabel("Break Points Faced")
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

tennis_data = pd.read_csv(r"C:\Users\monik\Downloads\tennis_ace_starting\tennis_ace_starting\tennis_stats.csv")

aces = tennis_data[['Aces']]
wins = tennis_data[['Wins']]
model2 = LinearRegression()
model2.fit(aces, wins)
wins_predict = model2.predict(aces)
plt.plot(aces, wins_predict, color="red")
plt.scatter(aces, wins, alpha=0.4)
plt.title("Wins vs. Aces with Regression Line")
plt.xlabel("Aces")
plt.ylabel("Wins")
plt.show()
```
