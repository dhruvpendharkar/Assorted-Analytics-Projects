```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from datetime import datetime as dt

data1 = pd.read_csv(r"C:\Users\monik\Downloads\Netflix+Stocks+Capstone\Netflix Stocks Capstone\DJI.csv")

data2= pd.read_csv(r"C:\Users\monik\Downloads\Netflix+Stocks+Capstone\Netflix Stocks Capstone\NFLX.csv")

data3 = pd.read_csv(r"C:\Users\monik\Downloads\Netflix+Stocks+Capstone\Netflix Stocks Capstone\NFLX_daily_by_quarter.csv")

date1 = data1['Date']
close1 = data1['Close']
date2 = data2['Date']
close2= data2['Close']
blank = []

fig = plt.figure(figsize=(10, 10))
ax1 = plt.subplot(2, 2, 1)
ax2 = plt.subplot(2, 2, 2)

ax1.plot(date1, close1)
ax1.set_xticks(blank)
ax1.set_title("Dow Jones Monthly Value")
ax2.plot(date2, close2)
ax2.set_xticks(blank)
ax2.set_title("Netflix Monthly Value")

plt.show()
```
