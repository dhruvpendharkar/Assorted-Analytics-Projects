```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv(r"C:/Users/monik/Blog Projects/NBA T25 Scorers/2021 NBA T25 Scorers.csv")
```

```python
plt.title("Load vs. Points per 75 possessions")
plt.scatter(df["Load"], df["PP/75"])
plt.ylabel("Points per 75")
plt.xlabel("Offensive Load")
plt.savefig("load_v_points2021.png")
```

```python
from sklearn.linear_model import LinearRegression
```

```python
load = df[["Load"]]
pp = df[["PP/75"]]
```

```python
model = LinearRegression()
model.fit(load, pp)
```

```python
pp_exp = model.predict(load)
```

```python
plt.title("Load vs. Points per 75 possessions")
plt.scatter(df["Load"], df["PP/75"])
plt.ylabel("Points per 75")
plt.xlabel("Offensive Load")
plt.plot(load, pp_exp, color="red")
plt.savefig("load_v_points2021reg.png")
```

```python
expected = []
for i in range(25):
    val = pp_exp[i][0]
    expected.append(val)
```

```python
df1 = pd.DataFrame()
df1["Player"] = df["BPM"]
df1["Expected"] = expected
df1["Actual"] = df["PP/75"]
df1["PPOE"] = df1["Actual"] - df1["Expected"]
```

```python
plt.rc('font', size=25)
plt.figure(figsize=(20, 20))
plt.bar(df1["Player"], df1["PPOE"])
plt.title("Points per 75 Possession Over Expected", fontsize=25)
plt.xlabel("Player", fontsize=2, weight="bold")
plt.xticks(fontsize=25)
plt.yticks(fontsize=15)
plt.ylabel("Points per 75 over expected", fontsize=20)
plt.xticks(rotation=75, fontsize=10)
plt.savefig("ppoe.png")
```
