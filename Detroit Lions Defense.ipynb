```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv(r"C:/Users/monik/Blog Projects/All NFL Advanced Data/NFL Advanced Defense Wks 1-3.csv")
```

```python
print(df.head())
```

```python
plt.figure(figsize=(20, 20))
plt.bar(df["Abbr"], df["Dropback EPA"], color="blue")
plt.title("EPA on Dropbacks for All NFL Teams (Weeks 1-3)", fontsize=15)
plt.ylabel("Dropback EPA/Play", fontsize=15)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.grid()
plt.savefig("Detroit Defense EPA.png")
```

```python
plt.figure(figsize=(20, 20))
plt.scatter(df["Dropback EPA"], df['Dropback SR'].str[:-1].astype(float), color=["red", "blue", "blue", "blue", "blue", "blue", "blue", "blue", "blue", "blue", "blue", "blue", "blue", "blue", "blue", "blue", "blue", "blue", "blue", "blue", "blue", "blue", "blue", "blue", "blue", "blue", "blue", "blue", "blue", "blue", "blue", "blue"])
xmax = max(df["Dropback EPA"]) + 0.1
xmin = min(df["Dropback EPA"]) - 0.1
plt.xlabel("Dropback EPA/play", fontsize=15)
plt.ylabel("Dropback Success Rate", fontsize=15)
plt.title("Dropback EPA vs Dropback Success Rate for All NFL Teams (Weeks 1-3)", fontsize=15)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.xlim(xmax, xmin)
plt.grid()
plt.savefig("Detroit Defense EPA vs SR.png")
```
