```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame()


df["Player"] = ["Z. Wilson", "M. Jones", "T. Lawrence", "T. Tagovailoa", "J. Burrow",
                "J. Herbert", "K. Murray", "D. Haskins", "D. Jones", "Baker Mayfield", 
                "Josh Allen", "Sam Darnold", "Josh Rosen", "Lamar Jackson", "Mitch Trubisky",
                "Patrick Mahomes", "Deshaun Watson", "Jared Goff", "Carson Wentz", "Paxton Lynch"]

df["first game passer rating"] = [82.9, 102.6, 70.1, 80.3, 66.1, 94.4, 75.2, 86.2, 112.7, 70.7, 63.3, 116.8, 88.5, 70.1, 60.1, 76.4, 75.9, 65.8, 101, 79.2]


df["rookie passer rating"] = [None, None, None, 87.1, 89.8, 98.3, 87.4, 76.1, 87.7, 93.7, 67.9, 77.6, 66.7, 84.5, 77.5, 76.4, 103, 63.6, 79.3, 81]
```

```python
df1 = df.sort_values("first game passer rating", ascending = False)
plt.figure(figsize=(20, 20))
ax = plt.axes()
plt.bar([i*2 for i in range(20)], df1["first game passer rating"], color = ["cyan", "cyan", "red", "cyan", "cyan", "cyan", "cyan", "red", "cyan", "cyan", "cyan", "cyan", "cyan", "cyan", "cyan", "red", "cyan", "cyan", "cyan", "cyan"], align = "center", width = 0.8)
plt.title("Rookie QB Debut Passer Ratings (2016-2021)", fontsize=20)
plt.xlabel("Player", fontweight="bold", fontsize=20)
plt.ylabel("Passer Rating", fontweight="bold", fontsize=20)
plt.xticks([i*2 for i in range(20)], df1["Player"], rotation=30, fontweight="bold", fontsize=15)
plt.yticks(fontsize=15)
ax.set_facecolor("black")
plt.savefig("rookie_qb_graph1.png")
```

```python
from sklearn.linear_model import LinearRegression
```

```python
d1 = []
d2 = []
for i in range(3, len(df["first game passer rating"])):
  val1 = df["first game passer rating"][i]
  val2 = df["rookie passer rating"][i]
  d1.append(val1)
  d2.append(val2)
```

```python
x = np.array(d1).reshape((-1, 1))
y = np.array(d2)
```

```python
model = LinearRegression()
model.fit(x, y)
```

```python
print(model.score(x, y))
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(d1, d2)
plt.xlabel("Rookie Debut Passer Rating", fontweight="bold", fontsize=10)
plt.ylabel("Rookie Season Passer Rating", fontweight="bold", fontsize=10)
plt.title("Rookie QB Debut Passer Ratings vs. Rookie Season Passer Ratings (2016-2020)", fontweight="bold", fontsize=10)
plt.savefig("rookie_qb_analysis.png")
```
